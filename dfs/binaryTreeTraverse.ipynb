```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
    def construct_binary_tree(self, root_list):
        def build_tree(index):
            if index >= len(root_list) or root_list[index] is None:
                return None
            node = TreeNode(root_list[index])
            node.left = build_tree(2 * index + 1)
            node.right = build_tree(2 * index + 2)
            return node

        return build_tree(0)


class Solution:
    def preorderTraversal(self, root: TreeNode):
        res = []
        def dfs(node):
            if node:
                res.append(node.val)
                dfs(node.left)
                dfs(node.right)
        
        dfs(root)
        return res
```

```python
# root_list = [1, None, 2, 3]
root_list = [1,2,5,3,4,None, None]
root = TreeNode().construct_binary_tree(root_list)
```

```python
Solution().preorderTraversal(root)
```

```python
# Definition for a binary tree node.
class Solution:
    def inorderTraversal(self, root: TreeNode):
        res = []
        def dfs(node):
            if node:
                dfs(node.left)
                res.append(node.val)
                dfs(node.right)
        
        dfs(root)
        return res
```

```python
root_list = [4,2,5,1,3,None, None]
root = TreeNode().construct_binary_tree(root_list)
```

```python
Solution().inorderTraversal(root)
```
